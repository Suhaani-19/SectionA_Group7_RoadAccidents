## Load Dataset

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("RTA Dataset.csv")
print("Shape:", df.shape)
print("Columns:", df.columns.tolist())
```

```python
df.head()
```

## Select Relevant Columns
Keep only required columns for focused analysis

```python
cols = [
    "Accident_severity",
    "Number_of_casualties",
    "Time",
    "Day_of_week",
    "Area_accident_occured",
    "Types_of_Junction",
    "Road_allignment",
    "Lanes_or_Medians",
    "Weather_conditions",
    "Light_conditions",
    "Road_surface_conditions",
    "Type_of_collision",
    "Vehicle_movement",
    "Cause_of_accident",
    "Age_band_of_driver",
    "Driving_experience",
    "Sex_of_driver",
    "Type_of_vehicle"
]

df = df[cols]

print("Shape after column selection:", df.shape)
```

## Standardize Missing Values

```python
missing_values = ["Unknown", "unknown", "NA", "-", ""]

for value in missing_values:
    df = df.replace(value, np.nan)

df.isnull().sum()
```

## Check Unique Values (IMPORTANT BEFORE CLEANING)

```python
for col in df.columns:
    print("\nColumn:", col)
    print("Unique Values:", df[col].nunique())
    print(df[col].unique())
```

## Fill Missing Values

```python
for col in df.columns:
    if df[col].dtype == "object":
        mode_value = df[col].mode()[0]
        df[col] = df[col].fillna(mode_value)
    else:
        median_value = df[col].median()
        df[col] = df[col].fillna(median_value)

df.isnull().sum()
```

## Clean Text Data

```python
for col in df.columns:
    if df[col].dtype == "object":
        df[col] = df[col].str.strip()
        df[col] = df[col].str.lower()
```

## Convert Data Types

```python
# Convert Number_of_casualties to numeric
df["Number_of_casualties"] = pd.to_numeric(df["Number_of_casualties"], errors="coerce")

# Convert Time to datetime
df["Time"] = pd.to_datetime(df["Time"], format="%H:%M:%S", errors="coerce")
```

```python
df.isna().sum()
```

## Remove Invalid Values

```python
# Remove negative casualties if any
df = df[df["Number_of_casualties"] >= 0]
# len(df[df["Number_of_casualties"] < 0]) = 0
```

## Remove Duplicates

```python
df = df.drop_duplicates()
df = df.reset_index(drop=True)

print("Final Shape:", df.shape)
```

# FEATURE ENGINEERING

## Extract Hour

```python
df["Hour"] = df["Time"].dt.hour
```

## Create Time Period

```python
def get_time_period(hour):
    if hour >= 0 and hour < 6:
        return "night"
    elif hour >= 6 and hour < 12:
        return "morning"
    elif hour >= 12 and hour < 18:
        return "afternoon"
    else:
        return "evening"

df["Time_period"] = df["Hour"].apply(get_time_period)
```

## Create Severity Score

```python
severity_map = {
    "slight injury": 1,
    "serious injury": 2,
    "fatal injury": 3
}

df["severity_score"] = df["Accident_severity"].map(severity_map)
```

## Create Weekend Flag

```python
df["is_weekend"] = df["Day_of_week"].isin(["saturday", "sunday"])
```
